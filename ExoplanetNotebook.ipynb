```python
import pandas as pd

# Load dataset
df = pd.read_csv('nasa_exoplanet_archive.csv')

# Filter by the criteria of radius, stellar flux (insolation) and temperature to find habitable environments
habitable = df[
    (df['pl_rade'] >= 0.5) & (df['pl_rade'] <= 2.0) &
    (df['pl_insol'] >= 0.35) & (df['pl_insol'] <= 1.5) &
    (df['st_teff'] >= 4000) & (df['st_teff'] <= 7000)
]

# Preview results
print(habitable[['pl_name', 'hostname', 'pl_rade', 'pl_insol', 'st_teff']])
```
